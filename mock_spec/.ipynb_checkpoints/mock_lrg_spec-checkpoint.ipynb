# 1. generate new coeff based upon LRG PCA-fitting

```python
import subprocess
import os
import copy

dir1 = os.getcwd()
os.chdir('./data')
dir2 = os.getcwd()

output = subprocess.Popen(['ls','-l'],stdout=subprocess.PIPE,shell=True)
lines = output.stdout.read().splitlines()
lines = [line.decode("utf-8") for line in lines]
print('number of hdf5 file',len(lines))
print(lines[1],lines[-1])
os.chdir(dir1)
```

```python
import h5py
import pickle
import gzip
import numpy as np

num_of_lrgs = 10000

ii = 0
tot_gal = 0
n_lrg = 0
newlines = ["{}/data/{}".format(dir1,item) for item in lines]
coeffs = []
count_lrgs = 0
for h5_file in newlines[:]:
    try:
        with h5py.File(h5_file, 'r') as f:
            arr0 = f['flux']
            flux_arr = arr0[:]
            
            arr1 = f['coeff']
            coeff_arr = arr1[:]
            
            res_ivar = f['res_ivar'] 
            res_ivar_arr = res_ivar[:]
            
            loglam = f['loglam']
            loglam_arr = loglam[:] 
            
            syn_flux = f['newsynflux']
            syn_flux_arr = syn_flux[:]
    except Exception as e: 
        print(e)
        print(h5_file)
    else:
        coeffs.append(coeff_arr)
        count_lrgs += coeff_arr.shape[0]
        if count_lrgs > num_of_lrgs:
            break

coeff_all = np.vstack(coeffs)[0:num_of_lrgs,:]
```

```python
from scipy import stats
values = coeff_all.T
kernel = stats.gaussian_kde(values)

new_coeff = kernel.resample(10000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import hist

fig, axs = plt.subplots(4,2,figsize=(14,20))
for ii,ax in enumerate(axs.reshape(-1)):
    if ii<7:
        hist(values[ii,:], bins='knuth', ax=ax, histtype='stepfilled', alpha=0.6, density=True,label='data');
        hist(new_coeff[ii,:], bins='knuth', ax=ax, histtype='stepfilled', alpha=0.6, density=True,label='KDE');
        ax.legend(loc=2,prop={'family':'serif','size': 12})
        ax.set_title('Coefficient{}'.format(ii))
plt.tight_layout()
plt.show()
```

# 2. port synthesis spectra generation from IDL to python

```python
#read template
#/data/inspur_disk03/userdir/caoxy/idl_tool/idlspec2d/templates/spEigenGal-55740.fits
from astropy.io import fits

temp_loc = '/data/inspur_disk03/userdir/caoxy/idl_tool/idlspec2d/templates/spEigenGal-55740.fits'
temp_spec = fits.getdata(temp_loc)
temp_header = fits.getheader(temp_loc)
ntemp, temp_npix = temp_spec.shape
temp_loglam = temp_header['COEFF0'] + temp_header['COEFF1']*np.arange(temp_npix)

#below code show four template spectra
plt.figure(figsize=(6,12))
plt.subplot(411)
plt.plot(10**temp_loglam,temp_spec[0,:])
plt.subplot(412)
plt.plot(10**temp_loglam,temp_spec[1,:])
plt.subplot(413)
plt.plot(10**temp_loglam,temp_spec[2,:])
plt.subplot(414)
plt.plot(10**temp_loglam,temp_spec[3,:])
plt.tight_layout()
plt.show()
```

```python
def interpol_template(temp_spec, temp_loglam, obj_loglam, obj_z):
    rest_loglam = obj_loglam - np.log10(1.0 + obj_z)
    ret_temp_spec = np.zeros((temp_spec.shape[0], rest_loglam.size))
    for ii in range(temp_spec.shape[0]):
        ret_temp_spec[ii,:] = np.interp(rest_loglam, temp_loglam, temp_spec[ii,:])
    return ret_temp_spec

#below code make the function works
'''
ret_temp_spec = interpol_template(temp_spec, temp_loglam, temp_loglam, 0.0)
for ii in range(ntemp):
    plt.figure(figsize=(4,2))
    plt.plot(10**temp_loglam, temp_spec[ii,:], alpha=0.5)
    plt.plot(10**temp_loglam, ret_temp_spec[ii,:], alpha=0.5)
    plt.show()
'''
```

```python
def poly_terms(x,npoly=3):
    ret = np.zeros((npoly,x.size))
    for ii in range(npoly):
        ret[ii,:] = x**ii
    return ret

def mock_spec_from_coeff(
    temp_spec, 
    temp_loglam, 
    obj_loglam, 
    obj_z, 
    coeffs, 
    npoly=3
    ):
    
    ntemp = temp_spec.shape[0]
    npix = obj_loglam.size
    obj_temp_spec = interpol_template(temp_spec, temp_loglam, obj_loglam, obj_z)
    
    #Normalize the templates
    for ii in range(ntemp):
        obj_temp_spec[ii,:] = obj_temp_spec[ii,:] / np.sqrt(np.mean(obj_temp_spec[ii,:]**2)) 
    
    #Add on some polynomial terms:
    pbase =  np.arange(npix,dtype='float')/(npix-1)
    polyset = poly_terms(pbase,npoly)

    temp_with_poly = np.vstack([obj_temp_spec,polyset])
    
    return np.dot(coeffs,temp_with_poly)
```

```python
#make sure mock spec works
with h5py.File(newlines[0], 'r') as f:
    arr0 = f['flux']
    flux_arr = arr0[:]
    
    arr1 = f['coeff']
    coeff_arr = arr1[:]
    
    res_ivar = f['res_ivar'] 
    res_ivar_arr = res_ivar[:]
    
    loglam = f['loglam']
    loglam_arr = loglam[:] 
    
    syn_flux = f['newsynflux']
    syn_flux_arr = syn_flux[:]
    
    zans = f['zans']
    zans_arr = zans[:]

ii=0
flux = flux_arr[ii,:]
loglam = loglam_arr[:]
z = zans_arr[ii][-9]
synflux = syn_flux_arr[ii,:]
coeff = coeff_arr[ii,:]

mock_spec = mock_spec_from_coeff(
    temp_spec, 
    temp_loglam, 
    loglam, 
    z, 
    coeff, 
    npoly=3
)

plt.figure(figsize=(15,3))
plt.plot(10**loglam,mock_spec)
plt.plot(10**loglam,synflux,'*',alpha=0.03)
plt.show()
```

# 3. use the new coeffs based on KDE to simulate spectra

```python
mock_spec = mock_spec_from_coeff(
    temp_spec, 
    temp_loglam, 
    loglam, 
    0.5, 
    new_coeff[:,0], 
    npoly=3
)

plt.figure(figsize=(15,3))
plt.plot(10**loglam,mock_spec)
plt.show()
```

# 4. Add noise using sky-fiber

```python
import subprocess
import os
import copy

dir1 = os.getcwd()
os.chdir('./sky_data')
dir2 = os.getcwd()

output = subprocess.Popen(['ls','-l'],stdout=subprocess.PIPE,shell=True)
lines = output.stdout.read().splitlines()
lines = [line.decode("utf-8") for line in lines]
print('number of hdf5 file',len(lines))
print(lines[1],lines[-1])
os.chdir(dir1)

import h5py
import pickle
import gzip
import numpy as np

num_of_sky = 10000

ii = 0
tot_gal = 0
n_lrg = 0
newlines = ["{}/sky_data/{}".format(dir1,item) for item in lines]
flux_sky = []
loglam_sky = []
count_sky = 0
for h5_file in newlines[:]:
    try:
        with h5py.File(h5_file, 'r') as f:
            arr0 = f['flux']
            flux_arr = arr0[:]
            
            
            arr1 = f['loglam']
            loglam_arr = arr1[:] 
            
    except Exception as e: 
        print(e)
        print(h5_file)
    else:
        flux_sky.append(flux_arr)
        loglam_sky.append(loglam_arr)
        count_sky += flux_arr.shape[0]
        if count_sky > num_of_sky:
            break

flux_sky_list = []
loglam_sky_list = []
for ii in range(len(flux_sky)):
    for jj in range(flux_sky[ii].shape[0]):
        flux_sky_list.append(flux_sky[ii][jj,:])
        loglam_sky_list.append(loglam_sky[ii])
```

```python
ind = np.random.randint(low=0, high=len(flux_sky_list), size=None)
noise = np.interp(loglam, loglam_sky_list[ind], flux_sky_list[ind])
mock_spec_with_noise = mock_spec + noise

plt.style.use('seaborn-paper')
font = {'family':'serif', 'size':15}
plt.rc('font', **font)

plt.figure(figsize=(12,17))
plt.subplot(411)
plt.plot(10**loglam,noise)
plt.title("One noise-realization from sky-fiber")
plt.subplot(412)
plt.plot(10**loglam,mock_spec)
plt.title("Ideal Mock LRG spec without noise")
plt.subplot(413)
plt.plot(10**loglam,mock_spec_with_noise)
plt.title("Mock LRG spec with noise")
plt.subplot(414)
plt.plot(10**loglam,mock_spec_with_noise)
plt.title("Mock LRG spec between 3600-5000")
plt.xlim(3600,5000)
plt.tight_layout()
plt.show()
```
